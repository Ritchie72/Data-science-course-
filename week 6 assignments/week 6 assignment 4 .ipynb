## Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is the process of mapping URLs to specific functions using decorators like @app.route(). It allows us to define the paths that users can access and helps organize how the web application responds to different requests.

## Q4.  Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: Company Name: ABC Corporation Location: India Contact Detail: 999-999-9999 

```python
from flask import Flask,jsonify

app = Flask(__name__)

@app.route('/')
def company_details():
    data = {
        'Company Name': 'ABC Corporation',
        'Location': 'India',
        'Contact Detail':'999-999-9999'
    }

    return jsonify(data)

@app.route('/welcome')
def welcome_msg():
    return 'Welcome to ABC corporation'


if __name__ == "__main__":
    app.run()
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

n Flask, the url_for() function is used for URL building by dynamically generating URLs based on the endpoint name.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/home')
def home():
    return "Welcome to the Home Page!"

with app.test_request_context():
    print(url_for('home'))  # Outputs: /home
```

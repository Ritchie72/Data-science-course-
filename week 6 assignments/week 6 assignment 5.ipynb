## Q1. Explain GET and POST methods

GET: Retrieves data from a server using parameters in the URL; typically used for requesting information without modifying data.

POST: Sends data to a server for processing; used for submitting data, such as form inputs, and can modify server data.

## Q2. Why is request used in Flask?

In Flask, request is used to access incoming data, such as form inputs or query parameters, from client requests. It allows handling of GET, POST, and other HTTP methods for processing user input and managing server interactions.

## Q3. Why is redirect() used in Flask?

In Flask, redirect() is used to send a user to a different URL, often after form submission or processing. It helps guide users to another route or page, such as after logging in or completing an action.

## Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are HTML files with placeholders for dynamic content, allowing separation of front-end and back-end code. The render_template() function is used to render these templates with data, generating complete HTML pages for the user.

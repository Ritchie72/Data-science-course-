## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
```

```python
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.xlabel('This is a X-data')
plt.ylabel('This is a Y-data')
plt.title('Random points')
```

```python
## Scatter plot is use to create a plot with random points
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

```python
## Subplot is used to draw multiple plots in a single diagram
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('plot 1')
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('plot 2')
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('plot 3')
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('plot 4')
```
